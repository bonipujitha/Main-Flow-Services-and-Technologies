```python
df.target.value_counts()
```

```python
#Plotting bar chart
df.target.value_counts().plot(kind= 'bar' , color=["orange" , "salmon"])
plt.title("Heart Disease Values")
plt.xlabel("1= Heart Disease, 0 = No Heart Disease")
plt.ylabel("Amount")
```
