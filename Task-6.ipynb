```python
df['Review Rating'].value_counts()
```

```python
df['Review Rating'].value_counts().plot(kind='bar', color=["purple", "yellow"])
plt.title("Amazon Product Review Sentiments")
plt.xlabel("Review Rating (1=Positive, 0=Negative)")
plt.ylabel("Count")
plt.show()
```

```python
df['Review Rating'].value_counts().plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%', colors=['lightskyblue', 'green'])
plt.title("Distribution of Amazon Product Review Sentiments")
plt.legend(["Positive", "Negative"])
plt.show()
```

```python
rating_counts = df['Helpful Review Count'].value_counts()
print(rating_counts)
```

```python
df['Helpful Review Count'].value_counts().plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%', colors=['green', 'yellow'])
plt.title("Distribution of Amazon Product Review Sentiments")
plt.legend(["Positive", "Negative"])
plt.show()
```

```python
cross_tab = pd.crosstab(df['Review Rating'], df['Helpful Review Count'])
print(cross_tab)
```
